1.What is a parameter?

ans
  
    In Machine Learning, a parameter is an internal variable that a model learns automatically during training to make predictions.

    For example, in Linear Regression, the equation is:
    y=mx+c

    Here:

    ùëö = slope ‚Üí model parameter (weight)

    ùëê = intercept ‚Üí model parameter (bias)
    
    During training, the algorithm adjusts these parameters to minimize the error between predicted and actual values.

2.What is correlation?
What does negative correlation mean?

ans

   
    Correlation measures the strength and direction of the relationship between two numerical variables.
    It ranges from -1 to +1:

    +1 ‚Üí perfect positive correlation (both increase together)

    0 ‚Üí no correlation

    -1 ‚Üí perfect negative correlation (one increases, other decreases)

    Example:

    X	Y
    1	2
    2	4
    3	6

    Here, X and Y have a correlation of +1 (perfectly proportional).

    A negative correlation means that when one variable increases, the other tends to decrease.
    Example:

    Number of hours spent watching TV vs. Exam scores.
    As TV hours increase, exam scores usually decrease ‚Äî this indicates a negative correlation.

5.What are continuous and categorical variables?

ans

    Continuous variables: Can take infinite numeric values within a range.
    Example: Height, weight, temperature.

    Categorical variables: Represent discrete categories or groups.
    Example: Gender (Male/Female), City (Delhi, Mumbai).

6.How do we handle categorical variables in Machine Learning? What are the common techniques?

ans


    Categorical data must be converted into numbers before feeding into ML models.

    Common techniques:

    Label Encoding ‚Äì Assigns each category a unique number.
    Example: {'Male': 0, 'Female': 1}

    One-Hot Encoding ‚Äì Creates binary columns for each category.
    Example: {'Red': [1,0,0], 'Blue': [0,1,0], 'Green': [0,0,1]}

    Ordinal Encoding ‚Äì Used when categories have order.
    Example: {'Low': 1, 'Medium': 2, 'High': 3}

15.What is causation? Explain difference between correlation and causation with an example.

 ans

    Causation: When one variable directly influences another.

    Correlation: Only shows a relationship, not cause and effect.

    Example:

    Ice cream sales and drowning cases are correlated (both rise in summer).
    ‚Üí But ice cream does not cause drowning ‚Üí correlation ‚â† causation.    


16.What is an Optimizer? What are different types of optimizers?

ans

    An optimizer updates the model parameters to reduce loss during training.

    Common types:

    Gradient Descent ‚Äì Updates weights using gradient of loss.

    Stochastic Gradient Descent (SGD) ‚Äì Updates weights per sample.

    Momentum ‚Äì Adds past gradient for faster convergence.

    Adam ‚Äì Adaptive learning rate + momentum (most popular).

    Example:

    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
